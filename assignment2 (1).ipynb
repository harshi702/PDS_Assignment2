```python
import pandas as pd
import datetime
import re
```

```python
data_path = '/content/drive/MyDrive/PDS_Assignment2/train.csv'
raw_data = pd.read_csv(data_path)
raw_data.head()
```

```python
raw_data.info()
```

```python
raw_data.isnull()
```

```python
raw_data.isnull().sum()
```

```python
#dropping new price column
raw_data.drop(columns=["New_Price"], inplace=True)
```

```python
#converting into float type
raw_data['Mileage'] = raw_data['Mileage'].str.extract('(\d+\.\d+)').astype(float)
raw_data['Engine'] = raw_data['Engine'].str.extract('(\d+)').astype(float)
raw_data['Power'] = raw_data['Power'].str.extract('(\d+\.\d+)').astype(float)
```

```python
# Impute missing values with the mean or median

raw_data['Mileage'].fillna(raw_data['Mileage'].median(), inplace=True)
raw_data['Engine'].fillna(raw_data['Engine'].median(), inplace=True)
raw_data['Power'].fillna(raw_data['Power'].median(), inplace=True)
raw_data["Seats"].fillna(raw_data["Seats"].median(), inplace=True)
```

```python
def numeric_value(text):
    if pd.notna(text):
        matches = re.findall(r'(\d+\.\d+|\d+)', str(text))
        if matches:
            return float(matches[0])
    return None

```

```python
#Task2
#Remove the units from some of the attributes and only keep the numerical value.
raw_data["Mileage"] = raw_data["Mileage"].apply(numeric_value)
raw_data["Engine"] = raw_data["Engine"].apply(numeric_value)
raw_data["Power"] = raw_data["Power"].apply(numeric_value)
```

```python
print(raw_data.describe())
```

```python

raw_data.isnull().sum()
```

```python
#Task3
# Perform one-hot encoding for 'Fuel_Type' and 'Transmission' columns
raw_data = pd.get_dummies(raw_data, columns=["Fuel_Type", "Transmission"])
print(raw_data)
```

```python
raw_data['Fuel_Type'].replace({'Diesel': 0, 'Petrol': 1, 'Electric': 2}, inplace=True)
raw_data['Transmission'].replace({'Manual': 0, 'Automatic': 1}, inplace=True)
```

```python
print(raw_data.head())
```

```python
#Task4
# Create one more feature and add this column to the dataset
current_year = datetime.datetime.now().year
raw_data['Current_age'] = current_year - raw_data['Year']
```

```python
print(raw_data.head())
```

```python
#Task5
#Perform select, filter, rename, mutate, arrange and summarize with group by operations
select_columns=raw_data[['Name','Location','Year','Kilometers_Driven','Price','Mileage']]
select_columns
```

```python
#renaming data
rename=select_columns.rename(columns={'Location':'Loc','Mileage':'new_Mileage'})
rename
```

```python
#filtering data
filter_data = rename[raw_data['Owner_Type'] == 'First']
filter_data
```

```python
#arranging data
arrange_data=filter_data.sort_values(by='new_Mileage')
arrange_data.head()
```

```python
#grouping and summarizing
summary=arrange_data.groupby('Year')['Price'].mean()
summary.head(5)
```
